```python
# fname = 'part1/logs/exper_1_setup.txt'
for i in range(1,8):
    content = []
    fname = 'part1/logs/exper_' + str(i) + '_setup.txt'
    dumpname = 'experiment'+ str(i) + '.html'
    dumpF = open(dumpname,'w+')
    dumpstring = ''
    dumpstring = dumpstring + '<!DOCTYPE html>\n<html lang="en">\n  <head>\n    <meta charset="utf-8">\n    <title>Image list prediction</title>\n  <style type="text/css">\n   img {\n     width:200px;\n   }\n  </style>\n  </head>\n  <body>\n\n<h2>Tushar Nimbhorkar, Diede Rusticus </h2>\n<h1>Settings</h1>\n<table>\n'
    
    with open(fname) as f:
        content = f.readlines()
        content = [x.strip() for x in content]
    vocab_size = int(content[1][11:14])
    descr_type = content[3][-9:]
    descr_step_size = int(content[4][-1:])
    fraction = float(content[5][-4:])
    train_P = int(content[6][-3:])
    train_N = 3*train_P
    bin_size = 8
    kernel = content[8][-6:]
    Airplane_AP = float(content[9][5:12])
    motor_AP = float(content[9][13:20])
    Faces_AP = float(content[9][21:28])
    car_AP = float(content[9][29:36])
    MAP = float(content[10][-7:])
    dumpstring = dumpstring + '<tr><th>SIFT step size</th><td>' + str(descr_step_size)+ ' px</td></tr>\n'
    dumpstring = dumpstring + '<tr><th>SIFT block sizes</th><td>' + str(bin_size) + ' pixels</td></tr>\n'
    dumpstring = dumpstring + '<tr><th>SIFT method</th><td>' + descr_type + '-SIFT</td></tr>\n'
    dumpstring = dumpstring + '<tr><th>Vocabulary size</th><td>'+ str(vocab_size) + ' words</td></tr>\n'
    dumpstring = dumpstring + '<tr><th>Vocabulary fraction</th><td>' + str(fraction) + '</td></tr>\n'
    dumpstring = dumpstring + '<tr><th>SVM training data</th><td>' + str(train_P) + ' positive, ' + str(train_N) + ' negative per class</td></tr>\n'
    dumpstring = dumpstring + '<tr><th>SVM kernel type</th><td>' + kernel +'</td></tr>\n'
    dumpstring = dumpstring + '</table>\n'
    dumpstring = dumpstring + '<h1>Prediction lists (MAP: '+ str(MAP) + ')</h1>\n'
    dumpstring = dumpstring + '<h3><font color="red">Following are the ranking lists for the four categories.</font></h3>\n'
    dumpstring = dumpstring + '<h3><font color="red">The length of each column is 200 (containing all test images).</font></h3>\n'
    dumpstring = dumpstring + '<table>\n'
    dumpstring = dumpstring + '<thead>\n'
    dumpstring = dumpstring + '<tr>\n'
    dumpstring = dumpstring + '<th>Airplanes (AP: ' + str(Airplane_AP) +')</th><th>Cars (AP: ' + str(car_AP) + ')</th><th>Faces (AP: ' + str(Faces_AP) + ')</th><th>Motorbikes (AP: ' + str(motor_AP) + ')</th>\n'
    dumpstring = dumpstring + '</tr>\n'
    dumpstring = dumpstring + '</thead>\n'
    dumpstring = dumpstring + '<tbody>\n'
#     for loop for images
    A_img = []
    C_img = []
    F_img = []
    M_img = []

    import csv
    with open('part1/logs/exper_' + str(i) + '_rank_airplanes.csv', 'rb') as f:
        reader = csv.reader(f)
        for row in reader:
            A_img.append(int(row[0]))
    with open('part1/logs/exper_' + str(i) + '_rank_cars.csv', 'rb') as f:
        reader = csv.reader(f)
        for row in reader:
            C_img.append(int(row[0]))
    with open('part1/logs/exper_' + str(i) + '_rank_faces.csv', 'rb') as f:
        reader = csv.reader(f)
        for row in reader:
            F_img.append(int(row[0]))
    with open('part1/logs/exper_' + str(i) + '_rank_motorbikes.csv', 'rb') as f:
        reader = csv.reader(f)
        for row in reader:
            M_img.append(int(row[0]))

    naming = ['Caltech4/ImageData/airplanes_test/img','Caltech4/ImageData/cars_test/img',
              'Caltech4/ImageData/faces_test/img','Caltech4/ImageData/motorbikes_test/img']

    naming_ranked_A = []
    naming_ranked_C = []
    naming_ranked_F = []
    naming_ranked_M = []
    for j in range(len(A_img)):
    #     ax =  '%.3d' % A[j]
        ax = A_img[j]
        if ax <= 50 :
            ax1 = ax
            ax1 =  '%.3d' % ax1
            temp_append = naming[0] + str(ax1) + '.jpg'
            naming_ranked_A.append(temp_append)
        if ax >50 and ax <= 100 :
            ax1 = ax - 50
            ax1 =  '%.3d' % ax1
            temp_append = naming[3] + str(ax1) + '.jpg'
            naming_ranked_A.append(temp_append)
        if ax >100 and ax <= 150 :
            ax1 = ax - 100
            ax1 =  '%.3d' % ax1
            temp_append = naming[2] + str(ax1) + '.jpg'
            naming_ranked_A.append(temp_append)
        if ax >150:
            ax1 = ax -150
            ax1 =  '%.3d' % ax1
            temp_append = naming[1] + str(ax1) + '.jpg'
            naming_ranked_A.append(temp_append)

    for j in range(len(C_img)):
    #     ax =  '%.3d' % A[j]
        ax = C_img[j]
        if ax <= 50 :
            ax1 = ax
            ax1 =  '%.3d' % ax1
            temp_append = naming[0] + str(ax1) + '.jpg'
            naming_ranked_C.append(temp_append)
        if ax >50 and ax <= 100 :
            ax1 = ax - 50
            ax1 =  '%.3d' % ax1
            temp_append = naming[3] + str(ax1) + '.jpg'
            naming_ranked_C.append(temp_append)
        if ax >100 and ax <= 150 :
            ax1 = ax - 100
            ax1 =  '%.3d' % ax1
            temp_append = naming[2] + str(ax1) + '.jpg'
            naming_ranked_C.append(temp_append)
        if ax >150:
            ax1 = ax -150
            ax1 =  '%.3d' % ax1
            temp_append = naming[1] + str(ax1) + '.jpg'
            naming_ranked_C.append(temp_append)

    for j in range(len(F_img)):
    #     ax =  '%.3d' % A[j]
        ax = F_img[j]
        if ax <= 50 :
            ax1 = ax
            ax1 =  '%.3d' % ax1
            temp_append = naming[0] + str(ax1) + '.jpg'
            naming_ranked_F.append(temp_append)
        if ax >50 and ax <= 100 :
            ax1 = ax - 50
            ax1 =  '%.3d' % ax1
            temp_append = naming[3] + str(ax1) + '.jpg'
            naming_ranked_F.append(temp_append)
        if ax >100 and ax <= 150 :
            ax1 = ax - 100
            ax1 =  '%.3d' % ax1
            temp_append = naming[2] + str(ax1) + '.jpg'
            naming_ranked_F.append(temp_append)
        if ax >150:
            ax1 = ax -150
            ax1 =  '%.3d' % ax1
            temp_append = naming[1] + str(ax1) + '.jpg'
            naming_ranked_F.append(temp_append)

    for j in range(len(M_img)):
    #     ax =  '%.3d' % A[j]
        ax = M_img[j]
        if ax <= 50 :
            ax1 = ax
            ax1 =  '%.3d' % ax1
            temp_append = naming[0] + str(ax1) + '.jpg'
            naming_ranked_M.append(temp_append)
        if ax >50 and ax <= 100 :
            ax1 = ax - 50
            ax1 =  '%.3d' % ax1
            temp_append = naming[3] + str(ax1) + '.jpg'
            naming_ranked_M.append(temp_append)
        if ax >100 and ax <= 150 :
            ax1 = ax - 100
            ax1 =  '%.3d' % ax1
            temp_append = naming[2] + str(ax1) + '.jpg'
            naming_ranked_M.append(temp_append)
        if ax >150:
            ax1 = ax -150
            ax1 =  '%.3d' % ax1
            temp_append = naming[1] + str(ax1) + '.jpg'
            naming_ranked_M.append(temp_append)
    
    for i in range(len(naming_ranked_A)):
        dumpstring = dumpstring + '<tr><td><img src="' + naming_ranked_A[i] + '" /></td><td><img src="' + naming_ranked_C[i] + '" /></td><td><img src="' + naming_ranked_F[i] + '" /></td><td><img src="' + naming_ranked_M[i] + '" /></td></tr>\n'
        
    dumpstring = dumpstring + '</tbody>\n'
    dumpstring = dumpstring + '</table>\n'
    dumpstring = dumpstring + '  </body>\n'
    dumpstring = dumpstring + '</html>\n'
    dumpF.write(dumpstring)
    dumpF.close()
    
    
    
```
